```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler #for Scaling the data mostly used for SVMs
from sklearn.svm import LinearSVC
```

```python
acc_data=pd.read_csv("/Users/rohithmovva/Downloads/HMP_Dataset/Featured_Data.csv", thousands = ',')
```

```python
np.random.seed(42)
shuffle_index = np.random.permutation(len(acc_data))
acc_data=acc_data.iloc[shuffle_index]
acc_data.head()
```

```python
#x=acc_data.iloc[:,2:23]
#y=(acc_data["HMPactivity"] == 'Climb_stairs').astype(np.float64)
#y=acc_data["HMPactivity"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(acc_data.iloc[:,2:23], acc_data["HMPactivity"],test_size=.2,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
y_train.unique()
```

```python
x_train.head()
```

```python
y_test.values
```

```python
## LinearSVC which uses One-Vs-Rest classification for Multiclass Classification
svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("linear_svc", LinearSVC(C=1, loss="hinge", random_state=42)),
    ])
```

```python
y_pred_t_linear=svm_clf.fit(x_train, y_train).predict(x_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
activity={0:"Climb_stairs",1:"Drink_glass",2:"Getup_bed",3:"Pour_water",4:"Sitdown_chair",5:"Standup_chair",6:"Walk"}
labels=["Climb_stairs","Drink_glass","Getup_bed","Pour_water","Sitdown_chair","Standup_chair","Walk"]
print(pd.DataFrame(confusion_matrix(y_train.values,y_pred_t_linear,labels=labels), index=activity, columns=activity))
print('Accuracy: ' + str(accuracy_score(y_train.values, y_pred_t_linear)))
```

```python
y_pred_linear=svm_clf.fit(x_train, y_train).predict(x_test)
```

```python
print(pd.DataFrame(confusion_matrix(y_test.values,y_pred_linear,labels=labels), index=activity, columns=activity))
print('Accuracy: ' + str(accuracy_score(y_test.values, y_pred_linear)))
```

```python
## SVC with Kernel Linear uses One-Vs-One Classification
from sklearn.svm import SVC
svm_clf_l = Pipeline([
        ("scaler", StandardScaler()),
        ("svc_l", SVC(kernel="linear",C=1, random_state=42)),
    ])
```

```python
y_pred_t_l=svm_clf_l.fit(x_train, y_train).predict(x_train)
```

```python
print(pd.DataFrame(confusion_matrix(y_train.values,y_pred_t_l,labels=labels), index=activity, columns=activity))
print('Accuracy: ' + str(accuracy_score(y_train.values, y_pred_t_l)))
```

```python
y_pred_l=svm_clf_l.fit(x_train, y_train).predict(x_test)
```

```python
print(pd.DataFrame(confusion_matrix(y_test.values,y_pred_l,labels=labels), index=activity, columns=activity))
print('Accuracy: ' + str(accuracy_score(y_test.values, y_pred_l)))
```

```python
poly_kernel_svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="poly", degree=2, coef0=1, C=1))
    ])
```

```python
y_pred_t_poly=poly_kernel_svm_clf.fit(x_train,y_train).predict(x_train)
print(pd.DataFrame(confusion_matrix(y_train.values,y_pred_t_poly,labels=labels), index=activity, columns=activity))
print('Accuracy: ' + str(accuracy_score(y_train.values, y_pred_t_poly)))
```

```python
y_pred_poly=poly_kernel_svm_clf.fit(x_train,y_train).predict(x_test)
```

```python
activity={0:"Climb_stairs",1:"Drink_glass",2:"Getup_bed",3:"Pour_water",4:"Sitdown_chair",5:"Standup_chair",6:"Walk"}
print(pd.DataFrame(confusion_matrix(y_test.values,y_pred_poly,labels=labels), index=activity, columns=activity))
print('Accuracy: ' + str(accuracy_score(y_test.values, y_pred_poly)))
```
